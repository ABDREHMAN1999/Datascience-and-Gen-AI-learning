```python
import sqlite3
```

```python
conn = sqlite3.connect("students.db")
cursor = conn.cursor()
cursor.execute("PRAGMA foreign_keys = ON;")
```

```python
cursor.execute("""
CREATE TABLE Data_Science(
    student_id INT PRIMARY KEY NOT NULL,
    name TEXT NOT NULL, 
    batch_id INT NOT NULL
)
""")
```

```python
student_1 = (1,"Abdul", 2)
```

```python
cursor.execute("""
INSERT INTO Data_Science (student_id , name, batch_id)
VALUES(?,?,?)          
""",student_1)
```

```python
n = int(input('Enter the number of students that want to enroll: '))
list_of_students = []
for i in range(n):
    student_id = int(input("Enter the student id: "))
    name = input("Enter the name: ")
    batch_id = int(input("Enter the batch id: "))
    student = (student_id, name, batch_id)
    list_of_students.append(student)
```

```python
for student in list_of_students:
    cursor.execute("""
INSERT INTO Data_Science (student_id , name, batch_id)
VALUES(?,?,?)
""",student)
```

```python
conn.commit()
```

```python
cursor.execute("""
               SELECT * FROM Data_Science
               """)
```
