```python
import os 
import psycopg2
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
USER = os.getenv("user")
PORT = os.getenv("port")
PASSWORD = os.getenv("password")
HOST = os.getenv("host")
DATABASE = os.getenv("database")
```

```python
conn = psycopg2.connect(
    host=HOST,
    port=PORT,
    database=DATABASE,
    user=USER,
    password=PASSWORD
)
cursor = conn.cursor()
```

```python

conn.rollback()
cursor.execute("""
SELECT * FROM "Employees"

""")
cursor.fetchall()

```

```python
conn.rollback()
emp_id = int(input("Enter Emp_ID"))
Dep_ID = int(input("Enter Dep_ID"))
grade_id = int(input("Enter grade_id"))
Salary = int(input("Enter Salary"))
name = input("Enter name: ")
val = (emp_id,Dep_ID, grade_id,Salary,name)


cursor.execute("""
INSERT INTO "Employees" (emp_id, "Dep_ID", grade_id, "Salary", name) values(%s,%s,%s,%s,%s)
""",val)
```

```python
conn.commit()
```
